# DIVE Entry (One Substantive Question)

## D — Define
**Substantive question:**  
<Example: “How does NOx change over time, and are there noticeable spikes that could indicate short-term events or data ingestion issues?”>

**Why it matters (operational relevance):**

**Metric(s) / output(s):**

- Time grain: <hour/day/week>

- Expected dashboard impact: <time-series KPI tile + trend line>



```python
# --- CHANGE THESE TO MATCH YOUR DATA ---
# The time_col and value_col will be defined after initial cleaning steps.

df2 = df.copy()

# Strip any leading/trailing whitespace from column names
df2.columns = df2.columns.str.strip()

# Combine 'Date' and 'Time' columns into a new 'Timestamp' column
# The format '%d/%m/%Y %H.%M.%S' matches the example data '10/03/2004;18.00.00'
df2['Timestamp'] = pd.to_datetime(df2['Date'] + ' ' + df2['Time'], errors='coerce', format='%d/%m/%Y %H.%M.%S')

# Define all potential pollutant/numeric columns that might need cleaning
pollutant_and_env_cols = [
    'CO(GT)', 'PT08.S1(CO)', 'NMHC(GT)', 'C6H6(GT)', 'PT08.S2(NMHC)',
    'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)',
    'T', 'RH', 'AH'
]

for col in pollutant_and_env_cols:
    if col in df2.columns:
        # Replace commas with dots for decimal conversion
        df2[col] = df2[col].astype(str).str.replace(',', '.', regex=False)
        # Replace -200 with NaN and convert to numeric, coercing errors
        df2[col] = pd.to_numeric(df2[col].replace('-200', np.nan), errors='coerce')

# Now set the time_col and value_col to the newly processed columns
time_col = "Timestamp"   # Use the combined and parsed timestamp column
value_col = "CO(GT)"       # Example: Carbon Monoxide (change as needed)

# Ensure the selected columns are not NaN and sort by time
df2 = df2.dropna(subset=[time_col, value_col]).sort_values(time_col)

# Optional: resample to daily trend (edit if you want hourly/weekly)
daily = (
    df2.set_index(time_col)[value_col]
       .resample("D")
       .agg(["mean", "median", "max", "count"])
       .reset_index()
)

daily.head()
```

```python
fig = px.line(
    daily,
    x=time_col,
    y="mean",
    title=f"Time Series Trend: {value_col} (Daily Mean)",
    hover_data={"median": True, "max": True, "count": True}
)

fig.update_xaxes(rangeslider_visible=True)
fig.update_layout(hovermode="x unified")
fig.show()
```
